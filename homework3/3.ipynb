```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib_inline
```

```python
import math


x = np.arange(0, 4 * np.pi, 0.1)
y = list(map(math.sin, x))
```

```python
plt.plot(x, y)
plt.show()
```

```python
import math
from scipy.misc import derivative

def function(x):
    return np.sin(x)

def der(x):
    return derivative(function, x)

y = np.linspace(-6, 6)
plt.plot(y, der(y), label='Derivative')
plt.plot(y, function(y), label='original')
plt.legend(loc='upper left')
plt.grid(True)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(0, 25, 1)
y = np.arange(0, 25, 1)
x, y = np.meshgrid(x, y)
z = x**2 - y**2
```

```python

plt.figure(figsize=(10, 10))
axes = plt.axes(projection='3d')
axes.view_init(-200, 50)
axes.plot_surface(x, y, z)
```

```python
x = 1
y = 1
u = 1
v = 0
plt.quiver(x, y, u, v)
plt.show()
```

```python
nx, ny = .5, .5
x = np.arange(-3, 5, nx)
y = np.arange(-3, 5, nx)
X, Y = np.meshgrid(x, y)
dy = X**2 - Y**2
dx = np.ones(dy.shape)
dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)

plt.quiver(X, Y, dxu, dyu, color='purple')
plt.show()
```

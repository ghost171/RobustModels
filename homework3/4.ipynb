```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.arange(0, 25, 1)
y = np.arange(0, 25, 1)
x, y = np.meshgrid(x, y)
z = x**2 - y**2
```

```python

plt.figure(figsize=(10, 10))
axes = plt.axes(projection='3d')
axes.view_init(-200, 50)
axes.plot_surface(x, y, z)
```

```python
x = 1
y = 1
u = 1
v = 0
plt.quiver(x, y, u, v)
plt.show()
```

```python
x = np.linspace(-5, 5, 18)
y = np.linspace(-5, 5, 18)

f = lambda x, y: x**2-y**2
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
dz = (2*X, -2*Y) #        ,   

normalized_dz0 = dz[0]/np.sqrt(Y**2+X**2)
normalized_dz1 = dz[1]/np.sqrt(X**2+Y**2)
```

```python
plt.quiver(X, Y, normalized_dz0, normalized_dz1, color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('  f(x,y)=x^2-y^2')
plt.show()
```

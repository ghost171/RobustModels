Нормальным называется распределение вероятностей, которое для одномерного случая задаётся функцией Гаусса.



Нормальное распределение зависит от 4-х параметров:

математическое ожидание — «центр тяжести» распределения;
дисперсия — степень разброса случайной величины относительно математического ожидания;
коэффициент асимметрии — параметр формы распределения, определяющий его симметрию относительно математического ожидания;
коэффициент эксцесса — параметр распределения, задающий «остроту» пика распределения.

$$
    f(x) = \frac{1}{(\sigma \sqrt(2\pi))}\frac{-\exp((x - \mu)^2}{(2\sigma^2)}, где - \mu - математическое ожидание, \sigma^2- дисперсия
$$

Смысл нормального распределения становится понятен из его формы. Наиболее вероятные значения случайной величины расположены вблизи его пика. По мере удаления от него, вероятность значений уменьшается и если значение расположено в «хвосте» распределения, то оно очень маловероятно.

Построим график плотности и функции нормального распределения

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python

plt.plot(x, stats.norm.cdf(x, mu, sigma))
plt.show()
```

Биноминальное распределение - распределение количества «успехов» в последовательности из n независимых случайных экспериментов, таких что вероятность «успеха» в каждом из них равна p.

Представим себе испытание с двумя возможными исходами: A и B, где, скажем, А условно означает «успех», дополнительное событие B – «неудачу».
Серию независимых испытаний такого рода с одной и той же вероятностью успеха р=Р(А) называют испытаниями Бернулли.

Примером может служить последовательное бросание монеты, в котором условно выпадение герба есть успех, а выпадение решетки – неудача.

Каждый исход n испытаний здесь можно описать цепочкой событий $\omega = (\omega_1, \ldots, \omega_n)$ , где $\omega_k$ = A или B соответственно означает успех или неудачу в k-м испытании,





Положим 1-p=q. Для независимых исходов $\omega_1, \ldots, \omega_n, $ в последовательности независимых испытаний вероятность их совместного осуществления есть $P(\omega)=P(\omega_1) \ldots P(\omega_n)$,
и для цепочки $\omega=(\omega_1, \ldots, \omega_n)$,  в которой имеется ровно k успехов и n-k неудач, получим:

$$
    P(\omega) = p^kq^{(n-k)}
$$



Получается вероятность для биноминального распределения:

$$
    P(\xi=k) = C_n^k p^k q^{n-k}
$$

Построим график плотности и функции распределения биномиального распределения

```python
import matplotlib.pyplot as plt
n = 6
p = 0.6
r_values = list(range(n + 1))
dist = [stats.binom.pmf(r, n, p) for r in r_values ]
plt.bar(r_values, dist)
plt.show()
```

```python
import matplotlib.pyplot as plt
n = 6
p = 0.6
r_values = list(range(n + 1))
dist = [stats.binom.cdf(r, n, p) for r in r_values ] 
plt.bar(r_values, dist)
plt.show()
```

Дискретная случайная величина имеет распределение Пуассона с параметром λ, если:

$$
    P(\xi=k) = \frac{\lambda^k}{k!}e^{-\lambda}
$$

Распределение Пуассона также называется распределением редких событий.

Распределение Пуассона - это дискретное распределение, являющееся одним из важных предельных случаев биномиального распределения.



Пуассоновская модель П() обычно описывает схему редких событий: при некоторых предположениях о характере процесса появления случайных событий число событий, происшедших за фиксированный промежуток времени или в фиксированной области пространства, часто подчиняется пуассоновскому распределению.


Распределение Пуассона моделирует случайную величину, представляющую собой число событий, произошедших за фиксированное время, при условии, что данные события происходят с некоторой фиксированной средней интенсивностью и независимо друг от друга.

Построим график плотности и функции распределения распределения Пуассона

```python
r_values = list(range(n + 1))
dist = [stats.poisson.pmf(k=r, mu=3) for r in r_values]
plt.bar(r_values, dist)
plt.show()
```

```python
r_values = list(range(n + 1))
dist = [stats.poisson.cdf(k=r, mu=3) for r in r_values]
plt.bar(r_values, dist)
plt.show()
```

Говорят, что случайная величина имеет непрерывное равномерное распределение на отрезке [a, b] $a, b \in R$, если ее плотность имеет вид:

$$
    \begin{equation*}
    f(x) = 
        \begin{cases}
            1/(b-a) &\text{, $x \in [a, b]$}\\
            0 &\text{, $x \notin [a, b]$}
        \end{cases}
    \end{equation*}
$$

Представьте себе известную игру в «рулетку». Её исход определяется положением случайной точки на окружности, указывающей финальное положение шарика, случайно брошенного на вращающееся колесо рулетки.

Очевидно, вероятность попадания этой случайной точки $\xi$ в тот или иной интервал не зависит от его расположения на окружности и пропорциональна длине интервала.

азвернув окружность (скажем, имеющую единичную длину) на отрезок (0, 1] действительной прямой, вероятность попадания точки $\xi$ в интервал (a, b] можно определить формулой


$$
    P(a \leq \xi \leq b) = \int\limits_a^b p(x)dx
    \begin{equation*}
    p(x) = 
        \begin{cases}
            1 &\text{, $0 \leq x \leq 1$}\\
            0 &\text{, $x<0, x>1$}
        \end{cases}
    \end{equation*}
$$

Построим график плотности и функции распределения равномерного распределения

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

a, b = 1, 5
size = 1000
```

```python
uniform_distribution = uniform(loc=a, scale=b)
x = np.linspace(uniform_distribution.ppf(0), uniform_distribution.ppf(1), size)
```

```python
pdf = uniform_distribution.pdf(x)
plt.plot(x, pdf, 'r-', lw=5, alpha=0.6, label='pdf')
```

```python
cdf = uniform_distribution.cdf(x)
plt.plot(x, cdf, 'k-', lw=2, label='cdf')
plt.legend(loc='best', frameon=False)
```

```python
plt.hist(uniform_distribution.rvs(size=size), density=True, histtype='stepfilled', alpha=0.2)
```

Мы говорим, что случайная величина $\xi$ имеет экспоненциальное (показательное) распределение, если:

$$
    \begin{equation*}
    P(\xi < x) = 
        \begin{cases}
            0 &\text{при  $x \leq 0$}\\
            1-e^{(-\lambda)x} &\text{при  $x > 0 $}
        \end{cases}
    \end{equation*}
$$

Пусть $\xi$ время ожидания события, тогда из формулы (0) следует, что вероятность того, что это событие наступит раньше x равна $1 - e^{- \lambda x}$. . Этот удобный формализм позволяет описывать моменты возникновения случайных событий. Параметр $\lambda$ оценивается на основе реальных данных. 

Плотность экспоненциального распределения имеет вид

$$
    \begin{equation*}
    p(x) = 
        \begin{cases}
            0 &\text{при  $x \leq 0$}\\
            \lambda e^{- \lambda x} &\text{при  $x > 0 $}
        \end{cases}
    \end{equation*}
$$

, где $\lambda > 0$ - положительная постоянная, называемая параметром экспоненциального распределения.
$$
    M \xi = \frac{1}{\lambda}
$$

$$
    D \xi = \frac{1}{\lambda ^ 2}
$$

Из этого следует:

$$
    \begin{equation*}
    P(\xi > x) = 
        \begin{cases}
            0 &\text{при  $x \leq 0$}\\
            \lambda e^{- \lambda x} &\text{при  $x > 0 $}
        \end{cases}
    \end{equation*}
$$



Построим график плотности и функции распределения экспоненциальное распределения

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, stats.expon.pdf(x, mu, sigma))
plt.show()
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
plt.plot(x, stats.expon.cdf(x, mu, sigma))
plt.show()
```
